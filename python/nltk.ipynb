## Types of Tokenize

1. BI-Grams (two consicutive words)
2. TRI-GRAMs (three consicutive words)
3. N-Grams

```python
from nltk.util import bigrams,trigrams,ngrams


string='my mame is the anurag and i wnat to go hone as soon as possible'

quote_tokens=word_tokenize(string)
```

```python
from nltk.util import bigrams,trigrams,ngrams


quote_bigram=list(bigrams(quote_tokens))
print(quote_bigram)
```

```python

quote_trigram=list(trigrams(quote_tokens))

print(quote_trigram)
```

```python
quote_n_gram=list(ngrams(quote_tokens,4))

print(quote_n_gram)
```

## Stemming

- normaltizig the word in to noranl form


##  Types of steming

1. PotterSteemmer
2. LancasterStemmer
3. Snowball Stemmer


```python
# PotterStemmer

from nltk.stem import PorterStemmer

stemmer=PorterStemmer()

print(stemmer.stem('running'))
print(stemmer.stem('eating'))
```

```python
print(stemmer.stem(AI))
```

```python
words_to_stem=['give','giving','given','gives','gave','thinking','thought','thoughts']

for word in words_to_stem:
    print(word,'->',stemmer.stem(word))
```

```python
# 2. LancasterStemmer

from nltk.stem import LancasterStemmer


lancaster_stemmer_steamer=LancasterStemmer()

words_to_stem=['give','giving','given','gives','gave','thinking','thought','thoughts']

for word in words_to_stem:
    print(word,'->',lancaster_stemmer_steamer.stem(word))
```

```python
# 2. Snowball Stemmer

from nltk.stem import SnowballStemmer


snowball_stemmer_steamer=SnowballStemmer('english')

words_to_stem=['give','giving','given','gives','gave','thinking','thought','thoughts']

for word in words_to_stem:
    print(word,'->',snowball_stemmer_steamer.stem(word))
```

## Lamatizaion

- It will give you complete words


```python
from nltk.stem import wordnet
from nltk.stem import WordNetLemmatizer

word_lem=WordNetLemmatizer()
```

```python
words_to_lem=['give','giving','given','gives','gave','thinking','thought','thoughts','playing','played','play','plays','played']

for word in words_to_lem:
    print(word, '-->', word_lem.lemmatize(word))
```

## Stopwords

- words that does not have much sementic meaning in a text.

- preposiotion (in,on,at)
- pronouns (i,you,he,she)
- conjunctions (and,or,but)

```python
from nltk.corpus import stopwords

stop_words=stopwords.words('english')

print(stop_words)
print('english stopwords --->',len(stop_words))
```

```python
stop_words=stopwords.words('french')

print(stop_words)
print('French stopwords --->',len(stop_words))
```

```python
stop_words=stopwords.words('german')

print(stop_words)
print('German stopwords --->',len(stop_words))
```

## Named Entity Recognizezation (NER) 

```python
# Named Entity Recognizezation (NER) is a process of extracting named entities from text. 
# Named entities are pre-defined categories such as person names, organizations, locations, medical codes, time expressions, quantities, monetary values, percentages, etc. 
# Named entities are important for many applications such as information extraction, question answering, and text summarization.

from nltk import ne_chunk
from nltk.tokenize import word_tokenize

ne_sent='us president live in the white house'
token=word_tokenize(ne_sent)

ne_tags=ne_chunk(token)

print(ne_tags)
```
