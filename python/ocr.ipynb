```python
img = cv2.imread(r"C:\Users\anura\Downloads\invoice-sample.jpg")

h, w, c = img.shape
boxes = pytesseract.image_to_boxes(img)
for b in boxes.splitlines():
    b = b.split(' ')
    x1=int(b[1])
    y1=int(b[2])
    p1=(x1,h-y1)

    x2=int(b[3])
    y2=int(b[4])
    p2=(x2,h-y2)

    img = cv2.rectangle(img,
                        (int(b[1]), h - int(b[2])), (int(b[3]), h - int(b[4])),
                        (0, 255, 0),
                        1)

cv2.imshow('img', img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
